```python
import numpy as nm
import pandas as pd
from numpy import nan
from numpy import isnan
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from math import pow
from sklearn.metrics import accuracy_score, mean_absolute_error

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import explained_variance_score
from sklearn.metrics import median_absolute_error
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn import preprocessing
from numpy import savetxt
import seaborn as sns
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

import matplotlib.pyplot as plt

# ---------------------------------
def classification_knn(x_train, y_train,x_test,y_test):
    print('\033[1m' + "K-Neighbors")
    print('\033[0m')
    neigh = KNeighborsClassifier(n_neighbors=3)
    # Train the algorithm
    neigh.fit(x_train, y_train)
    # predict the response
    predictionsKNN = neigh.predict(x_test)
    # evaluate accuracy
    print('KNeighbors accuracy : ', accuracy_score(y_test, predictionsKNN))
    knn_accuracy = accuracy_score(y_test, predictionsKNN)
    target_names = ['class 0', 'class 1']
    print(classification_report(y_test, predictionsKNN, target_names=target_names))

    print('ROC:', roc_auc_score(y_test, predictionsKNN))

    print('Variance:', explained_variance_score(y_test, predictionsKNN, multioutput='uniform_average'))

    mae = mean_absolute_error(y_test, predictionsKNN)
    print('Mean absolute error:', mae)

    med = median_absolute_error(y_test, predictionsKNN)
    print('Median absolute error', med)

    cm = metrics.confusion_matrix(y_test, predictionsKNN)
    print('Confusion matrix:')
    print(cm)

    print('\n')
    return predictionsKNN

def classification_nb(x_train, y_train,x_test,y_test):
    print('\033[1m' + "Naive bayes")
    print('\033[0m')
    gnb = GaussianNB()
    y_pred = gnb.fit(x_train, y_train).predict(x_test)
    print("Naive-Bayes accuracy : ", accuracy_score(y_test, y_pred, normalize=True))
    nb_accuracy = accuracy_score(y_test, y_pred, normalize=True)
    target_names = ['class 0', 'class 1']
    print(classification_report(y_test, y_pred, target_names=target_names))

    print('ROC:', roc_auc_score(y_test, y_pred))

    print('Variance:', explained_variance_score(y_test, y_pred, multioutput='uniform_average'))

    mae = mean_absolute_error(y_test, y_pred)
    print('Mean absolute error:', mae)

    med = median_absolute_error(y_test, y_pred)
    print('Median absolute error', med)

    cm = metrics.confusion_matrix(y_test, y_pred)
    print('Confusion matrix:')
    print(cm)

    print('\n')
    return y_pred

def classification_dt(x_train,y_train,x_test,y_test):
    print('\033[1m' + "Decision tree")
    print('\033[0m')
    classifier = DecisionTreeClassifier()
    classifier.fit(x_train, y_train)
    predictionsDT = classifier.predict(x_test)
    print('Decision tree  accuracy:', accuracy_score(y_test, predictionsDT))

    tree_values.append(accuracy_score(y_test, predictionsDT))

    target_names = ['class 0', 'class 1']
    print(classification_report(y_test, predictionsDT, target_names=target_names))

    print('ROC:', roc_auc_score(y_test, predictionsDT))

    print('Variance:', explained_variance_score(y_test, predictionsDT, multioutput='uniform_average'))

    mae = mean_absolute_error(y_test, predictionsDT)
    print('Mean absolute error:', mae)

    med = median_absolute_error(y_test, predictionsDT)
    print('Median absolute error', med)

    cm = metrics.confusion_matrix(y_test, predictionsDT)
    print('Confusion matrix:')
    print(cm)

    print('\n')
    return predictionsDT

def classification_lr(x_train,y_train,x_test,y_test):
    print('\033[1m' + "Logistic regression")
    print('\033[0m')
    logisticRegr = LogisticRegression()
    logisticRegr.fit(x_train, y_train)
    predictionsLR = logisticRegr.predict(x_test)
    score = logisticRegr.score(x_test, y_test)
    print('Logistic regression:', score)
    print('Logistic regression accuracy:', accuracy_score(y_test, predictionsLR))
    logistic_regresion_values.append(accuracy_score(y_test, predictionsLR))

    target_names = ['class 0', 'class 1']
    print(classification_report(y_test, predictionsLR, target_names=target_names))

    print('ROC:', roc_auc_score(y_test, predictionsDT))

    print('Variance:', explained_variance_score(y_test, predictionsLR, multioutput='uniform_average'))

    mae = mean_absolute_error(y_test, predictionsLR)
    print('Mean absolute error:', mae)

    med = median_absolute_error(y_test, predictionsLR)
    print('Median absolute error', med)

    cm = metrics.confusion_matrix(y_test, predictionsLR)
    print('Confusion matrix:')
    print(cm)

    print('\n')
    return predictionsLR


dataset1 = pd.read_csv('phishing.csv')

# skaliranje
min_max_scaler = preprocessing.MinMaxScaler()
x_minmax = min_max_scaler.fit_transform(dataset1.values)

# savetxt('phishing2.csv', x_minmax,fmt='%.2f', delimiter=',')

dataset = pd.read_csv('phishing2.csv')

# provera nedostajuceih vrednosti
print('Broj nedostajucih vrednosti: ', dataset.isnull().sum().sum())
print('\n')

tree_values = []
logistic_regresion_values = []

# *****************************************************************************************************************
# bez metoda
print('\033[1m' + 'Bez redukcije atributa')
print('\033[0m')

array = dataset.values
x = array[:, 0:30]
y = array[:, -1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=1)

# klasifikatori
# --------------------------------
# K-Neighbors
predictionsKNN=classification_knn(x_train,y_train,x_test,y_test)

# ----------------------------------
# Naive bayes
y_pred=classification_nb(x_train,y_train,x_test,y_test)
nb_accuracy = accuracy_score(y_test, y_pred, normalize=True)
# ------------------------------------
# Decision tree
predictionsDT=classification_dt(x_train,y_train,x_test,y_test)
dt_accuracy=accuracy_score(y_test, predictionsDT)
# ----------------------------------
# Logistic regression
predictionsLR=classification_lr(x_train,y_train,x_test,y_test)
lr_accuracy=accuracy_score(y_test, predictionsLR)

recall_knn = recall_score(y_test, predictionsKNN)
f1_knn = f1_score(y_test, predictionsKNN)

recall_nb = recall_score(y_test, y_pred)
f1_nb = f1_score(y_test, y_pred)

recall_dt = recall_score(y_test, predictionsDT)
f1_dt = f1_score(y_test, predictionsDT)

recall_lr = recall_score(y_test, predictionsLR)
f1_lr = f1_score(y_test, predictionsLR)

# ***********************************************************************************************************************

# metoda Univariate Selection
print('\033[1m' + 'Metoda Univariate Selection')
print('\033[0m')

array = dataset.values
x = array[:, 0:30]
y = array[:, -1]

sel = SelectKBest(score_func=chi2, k=20)
sel.fit(x, y)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=1)

# Transform (remove low variance features)
x_train = sel.transform(x_train)
x_test = sel.transform(x_test)

# klasifikatori
# --------------------------------
# K-Neighbors
predictionsKNN=classification_knn(x_train,y_train,x_test,y_test)
knn_accuracy_US = accuracy_score(y_test, predictionsKNN)
# ----------------------------------
# Naive bayes
y_pred=classification_nb(x_train,y_train,x_test,y_test)
nb_accuracy_US= accuracy_score(y_test, y_pred, normalize=True)
# ------------------------------------
# Decision tree
predictionsDT=classification_dt(x_train,y_train,x_test,y_test)
dt_accuracy_US=accuracy_score(y_test, predictionsDT)
# ----------------------------------
# Logistic regression
predictionsLR=classification_lr(x_train,y_train,x_test,y_test)
lr_accuracy_US=accuracy_score(y_test, predictionsLR)

recall_knn_US = recall_score(y_test, predictionsKNN)
f1_knn_US = f1_score(y_test, predictionsKNN)

recall_nb_US = recall_score(y_test, y_pred)
f1_nb_US = f1_score(y_test, y_pred)

recall_dt_US = recall_score(y_test, predictionsDT)
f1_dt_US = f1_score(y_test, predictionsDT)

recall_lr_US = recall_score(y_test, predictionsLR)
f1_lr_US = f1_score(y_test, predictionsLR)

# ***********************************************************************************************************************
# metoda Feature Importance
print('\033[1m' + 'Metoda Feature Importance')
print('\033[0m')

from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt

model = ExtraTreesClassifier()
model.fit(x, y)
print(model.feature_importances_)  # use inbuilt class feature_importances of tree based classifiers

dataAtr = dataset[["double_slash_redirecting", 'Links_pointing_to_page', 'Links_in_tags',
                   'having_IP_Address', 'URL_Length', 'DNSRecord', 'Shortining_Service',
                   'Abnormal_URL', 'Redirect', 'HTTPS_token', 'Result']]
# print(dataAtr)


dataAtrValues = dataAtr.values
# train i test
x, y = dataAtrValues[:, :-1], dataAtrValues[:, -1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30)

# klasifikatori
# --------------------------------
# K-Neighbors
predictionsKNN=classification_knn(x_train,y_train,x_test,y_test)
knn_accuracy_FI = accuracy_score(y_test, predictionsKNN)

# ----------------------------------
# Naive bayes
y_pred=classification_nb(x_train,y_train,x_test,y_test)
nb_accuracy_FI = accuracy_score(y_test, y_pred, normalize=True)
# ------------------------------------
# Decision tree
predictionsDT=classification_dt(x_train,y_train,x_test,y_test)
dt_accuracy_FI=accuracy_score(y_test, predictionsDT)
# ----------------------------------
# Logistic regression
predictionsLR=classification_lr(x_train,y_train,x_test,y_test)
lr_accuracy_FI=accuracy_score(y_test, predictionsLR)

recall_knn_FI = recall_score(y_test, predictionsKNN)
f1_knn_FI = f1_score(y_test, predictionsKNN)

recall_nb_FI = recall_score(y_test, y_pred)
f1_nb_FI = f1_score(y_test, y_pred)

recall_dt_FI = recall_score(y_test, predictionsDT)
f1_dt_FI = f1_score(y_test, predictionsDT)

recall_lr_FI = recall_score(y_test, predictionsLR)
f1_lr_FI = f1_score(y_test, predictionsLR)



# ***********************************************************************************************************************
# metoda Correlation Matrix with Heatmap
# get correlations of each features in dataset
print('\033[1m' + 'Metoda Correlation Matrix with Heatmap')
print('\033[0m')

corrmat = dataset.corr()
top_corr_features = corrmat.index

# Correlation with output variable
cor_target = abs(corrmat["Result"])
# Selecting highly correlated features
relevant_features = cor_target[cor_target > 0.3]
# print(relevant_features)

# plot heat map
# g=sns.heatmap(dataset[top_corr_features].corr(),annot=True,cmap="RdYlGn")

dataAtr = dataset[["Prefix_Suffix", 'SSLfinal_State', 'URL_of_Anchor',
                   'web_traffic', 'Result']]
# print(dataAtr)

dataAtrValues = dataAtr.values
# train i test
x, y = dataAtrValues[:, :-1], dataAtrValues[:, -1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30)

# klasifikatori
# --------------------------------
# K-Neighbors
predictionsKNN=classification_knn(x_train,y_train,x_test,y_test)
knn_accuracy_CH = accuracy_score(y_test, predictionsKNN)
# ----------------------------------
# Naive bayes
y_pred=classification_nb(x_train,y_train,x_test,y_test)
nb_accuracy_CH = accuracy_score(y_test, y_pred, normalize=True)
# ------------------------------------
# Decision tree
predictionsDT=classification_dt(x_train,y_train,x_test,y_test)
dt_accuracy_CH=accuracy_score(y_test, predictionsDT)
# ----------------------------------
# Logistic regression
predictionsLR=classification_lr(x_train,y_train,x_test,y_test)
lr_accuracy_CH=accuracy_score(y_test, predictionsLR)

recall_knn_CM = recall_score(y_test, predictionsKNN)
f1_knn_CM = f1_score(y_test, predictionsKNN)

recall_nb_CM = recall_score(y_test, y_pred)
f1_nb_CM = f1_score(y_test, y_pred)

recall_dt_CM = recall_score(y_test, predictionsDT)
f1_dt_CM = f1_score(y_test, predictionsDT)

recall_lr_CM = recall_score(y_test, predictionsLR)
f1_lr_CM = f1_score(y_test, predictionsLR)

#####################################################
# grafici

# Decision tree
bars = (
'Without Feature Selection', 'Univariate Selection', 'Feature Importance', 'Correlation Matrix with Heatmap')

barWidth = 0.3
r1 = nm.arange(len(tree_values))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

plt.rcParams.update({'font.size': 11})
f, ax = plt.subplots(figsize=(12, 6))
ax.set_xlabel('Feature Selection Techniques')
ax.set_title('Decision tree classifier')

# recall
bars2 = (recall_dt, recall_dt_US, recall_dt_FI, recall_dt_CM)

# f measure
bars3 = (f1_dt, f1_dt_US, f1_dt_FI, f1_dt_CM)

# Create bars
plt.bar(r1, tree_values, width=0.3, edgecolor='black', capsize=7, label='accuracy')

# Create cyan bars
plt.bar(r2, bars2, width=barWidth, color='cyan', edgecolor='black', capsize=7, label='recall')

# Create cyan bars
plt.bar(r3, bars3, width=barWidth, color='#6c3376', edgecolor='black', capsize=7, label='F measure')

# Create names on the x-axis
plt.xticks([r + barWidth for r in range(len(tree_values))], bars)
plt.ylabel('Values')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show graphic
plt.show()

# Logistic regression
bars = (
'Without Feature Selection', 'Univariate Selection', 'Feature Importance', 'Correlation Matrix with Heatmap')

barWidth = 0.3
r1 = nm.arange(len(logistic_regresion_values))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

f, ax = plt.subplots(figsize=(12, 6))
ax.set_xlabel('Feature Selection Techniques')
ax.set_title('Logistic regression')

# recall
bars2 = (recall_lr, recall_lr_US, recall_lr_FI, recall_lr_CM)

# f measure
bars3 = (f1_lr, f1_lr_US, f1_lr_FI, f1_lr_CM)

# Create bars
plt.bar(r1, logistic_regresion_values, width=0.3, edgecolor='black', capsize=7, label='accuracy')

# Create cyan bars
plt.bar(r2, bars2, width=barWidth, color='cyan', edgecolor='black', capsize=7, label='recall')

# Create cyan bars
plt.bar(r3, bars3, width=barWidth, color='#6c3376', edgecolor='black', capsize=7, label='F measure')

# Create names on the x-axis
plt.xticks([r + barWidth for r in range(len(logistic_regresion_values))], bars)
plt.ylabel('Values')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show graphic
plt.show()

# kNN
knn_accuracy = accuracy_score(y_test, predictionsKNN)
y_KNN = (knn_accuracy, knn_accuracy_US, knn_accuracy_FI, knn_accuracy_CH)

bars = (
'Without Feature Selection', 'Univariate Selection', 'Feature Importance', 'Correlation Matrix with Heatmap')

barWidth = 0.3
r1 = nm.arange(len(y_KNN))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

f, ax = plt.subplots(figsize=(12, 6))
ax.set_xlabel('Feature Selection Techniques')
ax.set_title('Logistic regression')

# recall
bars2 = (recall_knn, recall_knn_US, recall_knn_FI, recall_knn_CM)

# f measure
bars3 = (f1_knn, f1_knn_US, f1_knn_FI, f1_knn_CM)

# Create bars
plt.bar(r1, y_KNN, width=0.3, edgecolor='black', capsize=7, label='accuracy')

# Create cyan bars
plt.bar(r2, bars2, width=barWidth, color='cyan', edgecolor='black', capsize=7, label='recall')

# Create cyan bars
plt.bar(r3, bars3, width=barWidth, color='#6c3376', edgecolor='black', capsize=7, label='F measure')

# Create names on the x-axis
plt.xticks([r + barWidth for r in range(len(y_KNN))], bars)
plt.ylabel('Values')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show graphic
plt.show()

# NaiveBayes
y_NB = (nb_accuracy, nb_accuracy_US, nb_accuracy_FI, nb_accuracy_CH)

bars = (
'Without Feature Selection Techniques', 'Univariate Selection', 'Feature Importance', 'Correlation Matrix with Heatmap')

barWidth = 0.3
r1 = nm.arange(len(y_NB))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

f, ax = plt.subplots(figsize=(12, 6))
ax.set_xlabel('Feature Selection Techniques')
ax.set_title('Logistic regression')

# recall
bars2 = (recall_nb, recall_nb_US, recall_nb_FI, recall_nb_CM)

# f measure
bars3 = (f1_nb, f1_nb_US, f1_nb_FI, f1_nb_CM)

# Create bars
plt.bar(r1, y_NB, width=0.3, edgecolor='black', capsize=7, label='accuracy')

# Create cyan bars
plt.bar(r2, bars2, width=barWidth, color='cyan', edgecolor='black', capsize=7, label='recall')

# Create cyan bars
plt.bar(r3, bars3, width=barWidth, color='#6c3376', edgecolor='black', capsize=7, label='F measure')

# Create names on the x-axis
plt.xticks([r + barWidth for r in range(len(y_NB))], bars)
plt.ylabel('Values')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show graphic
plt.show()

# best method
y_best = (knn_accuracy_US * 100, nb_accuracy_US * 100, dt_accuracy_US * 100, lr_accuracy_US * 100)
x_best = ('KNN', 'Naive Bayes', 'Decision tree', 'Logistic regression')
y_pos_best = nm.arange(len(x_best))

plt.figure(figsize=(7, 3))
plt.bar(y_pos_best, y_best, width=0.5)
plt.xticks(y_pos_best, x_best)
plt.title('Univariate Selection')
plt.xlabel("Classificator")
plt.ylabel("Accuracy")

plt.show()
######################***************
from sklearn.ensemble import VotingClassifier

# knn and decision tree
classificator_DT = DecisionTreeClassifier()
classificator_KNN = KNeighborsClassifier(n_neighbors=3)

classificator_KNN_DT = VotingClassifier(estimators=[('DT', classificator_DT), ('KNN', classificator_KNN)],
                                        voting='soft',
                                        weights=[1, 1])

score_KNN_DT = classificator_KNN_DT.fit(x_train, y_train)
predictionsKNN_DT = score_KNN_DT.predict(x_test)
accuracyKNN_DT = accuracy_score(y_test, predictionsKNN_DT)
# print ('KNN and Decision tree accuracy:',accuracyKNN_DT)

# KNN and logic regresion
classificator_LR = LogisticRegression()
classificator_KNN = KNeighborsClassifier(n_neighbors=3)

classificator_KNN_LR = VotingClassifier(estimators=[('LR', classificator_LR), ('KNN', classificator_KNN)],
                                        voting='soft',
                                        weights=[1, 1])

score_KNN_LR = classificator_KNN_LR.fit(x_train, y_train)
predictionsKNN_LR = score_KNN_LR.predict(x_test)
accuracyKNN_LR = accuracy_score(y_test, predictionsKNN_LR)
# print ('KNN and Logistic regression accuracy:',accuracyKNN_LR)

# Logistic regression and DT
classificator_LR = LogisticRegression()
classificator_DT = DecisionTreeClassifier()

classificator_DT_LR = VotingClassifier(estimators=[('LR', classificator_LR), ('DT', classificator_DT)],
                                       voting='soft',
                                       weights=[1, 1])

score_DT_LR = classificator_DT_LR.fit(x_train, y_train)
predictionsDT_LR = score_DT_LR.predict(x_test)
accuracyDT_LR = accuracy_score(y_test, predictionsDT_LR)
# print ('Logistic regression and Decision tree accuracy:',accuracyDT_LR)

# LR,DT and KNN
classificator_LR = LogisticRegression()
classificator_DT = DecisionTreeClassifier()
classificator_KNN = KNeighborsClassifier(n_neighbors=3)

classificator_DT_LR_KNN = VotingClassifier(
    estimators=[('LR', classificator_LR), ('DT', classificator_DT), ('KNN', classificator_KNN)],
    voting='soft',
    weights=[1, 1, 1])

score_DT_LR_KNN = classificator_DT_LR_KNN.fit(x_train, y_train)
predictionsDT_LR_KNN = score_DT_LR_KNN.predict(x_test)
accuracyDT_LR_KNN = accuracy_score(y_test, predictionsDT_LR_KNN)
# print ('Logistic regression,Decision tree and KNN accuracy:',accuracyDT_LR_KNN)


# grafik

y_comb = (accuracyKNN_DT * 100, accuracyKNN_LR * 100, accuracyDT_LR * 100, accuracyDT_LR_KNN * 100)
x_comb = ('KNN and DT', 'KNN and LR', 'DT and LR', 'DT, LR and KNN')
y_pos_comb = nm.arange(len(x_best))

plt.rcParams.update({'font.size': 11})

plt.figure(figsize=(7, 3))
plt.bar(y_pos_comb, y_comb, width=0.5)
plt.xticks(y_pos_comb, x_comb)
plt.title('Classifier combinations')
plt.xlabel("Classificator")
plt.ylabel("Accuracy")

plt.show()
```
